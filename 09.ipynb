```python
%matplotlib inline 
from sklearn import __version__ as sklearn_version
from distutils.version import LooseVersion
from IPython.display import Image 
from sklearn import datasets
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.kernel_ridge import KernelRidge
from sklearn.svm import SVR
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import pandas as pd
# df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data',header=None)
# df.tail()
# # print(df);
# ab = df.iloc[0:11, 0:5];
# print(ab);
boston = datasets.load_boston()
boston=np.column_stack((boston.data,boston.target))



for i in range(0,14):
    for j in range(i+1,14):
        X=boston[ : , [i,j]]
        plt.xlabel(str(i) +' th attribute')
        plt.ylabel(str(j) +' th attribute')
        plt.scatter(X[:, 0],X[:,1])
        plt.show()
```

```python
# 0th and 6th ,0th and 7th,0th and 13th, 
# 1th and 2nd ,1st and 4th,1st and 6th, 1st and 12th, 
# 2nd and 4th,2nd and 7th,
# 4th and 6th,4th and 7th,
# 5th and 12th,,5th and 13th, 
# 6th and 7th ,6th and 12th ,6th and 13th 
# 7th and 12th, 7th and 13th
# 12th and 13th have pattern
# therefore, i thought 6th and 7th column of the data will be predict by other data in the dataset
```

```python
X=np.column_stack((boston[:,0:6],boston[:,7:14]))
# print(X)
y=boston[:, 6]
# print(y.shape)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1)
```

```python
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))
```

```python
lr = LinearRegression()
lr.fit(X_train_std, y_train)
print("LinearRegression", lr.score(X_combined_std, y_combined))
```

```python
rr= Ridge()
rr.fit(X_train_std, y_train)
print("Ridge",rr.score(X_combined_std, y_combined))
```

```python
krr=KernelRidge()
krr.fit(X_train_std, y_train)
print("Kernelridge",krr.score(X_combined_std, y_combined))
```

```python
svr=SVR()
svr.fit(X_train_std, y_train)
print("SVR",svr.score(X_combined_std, y_combined))
```

```python
lasso=Lasso()
lasso.fit(X_train_std, y_train)
print("lasso",lasso.score(X_combined_std, y_combined))
```

```python
ela=ElasticNet()
ela.fit(X_train_std, y_train)
print("ElasticNet",ela.score(X_combined_std, y_combined))
```

```python
tree=DecisionTreeRegressor()
tree.fit(X_train_std, y_train)
print("DecisionTreeRegressor",tree.score(X_combined_std, y_combined))
```

```python
forest=RandomForestRegressor()
forest.fit(X_train_std, y_train)
print("RandomForestRegressor",forest.score(X_combined_std, y_combined))
```
